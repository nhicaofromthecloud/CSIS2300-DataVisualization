```python
import pandas as pd

# Read the dataset (replace 'your_dataset.csv' with the actual file name)
data = pd.read_csv('vancouver.csv')

data
```

```python
# Remove commas in columns 1, 2, and 3
for col in range(1, 4):
    data.iloc[:, col] = data.iloc[:, col].str.replace(',', '').astype(float)
```

```python
# Convert the object type columns to int
cols_to_convert = [1, 2, 3]
for col in cols_to_convert:
    data.iloc[:, col] = pd.to_numeric(data.iloc[:, col], errors='coerce').fillna(0).astype(int)

# Check the converted data types
print(data.dtypes)

data
```

```python
# Get the value in the first row of column 1 (assuming 0-based indexing)
first_value = data.iloc[0, 1]

# Calculate the scaling factor
scaling_factor = 100 / first_value

# Scale the dataset columns 1, 2, and 3
for col in range(1, 4):
    data.iloc[:, col] = data.iloc[:, col] * scaling_factor

# Check the scaled first row in column 1 (it should be 100)
print("Scaled first row in column 1:", data.iloc[0, 1])

```

```python
# Save the scaled dataset to a new file 
data.to_csv('vancouver_scaled_dataset.csv', index=False)
```
